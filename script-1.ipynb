```python
import pandas as pd
```

```python
df = pd.read_excel("dataset/Pokemon.xlsx")
df.head()
```

```python
# Correlaciones entre variables numéricas
print("\n" + "=" * 60)
print("MATRIZ DE CORRELACIONES")
print("=" * 60)
numeric_df = df.select_dtypes(include=['number'])
if len(numeric_df.columns) > 0:
    correlation = numeric_df.corr()
    print(correlation.round(3))
else:
    print("No hay variables numéricas en el dataset")
```

```python
# Variables categóricas
print("\n" + "=" * 60)
print("ANÁLISIS DE VARIABLES CATEGÓRICAS")
print("=" * 60)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(f"  Valores únicos: {df[col].nunique()}")
    print(f"  Valores más comunes:")
    print(df[col].value_counts().head().to_string())
```

```python
# Estadísticas descriptivas
print("\n" + "=" * 60)
print("ESTADÍSTICAS DESCRIPTIVAS")
print("=" * 60)
print(df.describe())
```

```python
# Valores faltantes
print("\n" + "=" * 60)
print("ANÁLISIS DE VALORES FALTANTES")
print("=" * 60)
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Columna': missing.index,
    'Valores_Faltantes': missing.values,
    'Porcentaje': missing_pct.values
}).sort_values('Valores_Faltantes', ascending=False)
print(missing_df.to_string(index=False))
```

```python
# Información básica del dataset
print("=" * 60)
print("INFORMACIÓN GENERAL DEL DATASET")
print("=" * 60)
print(f"\nDimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
print(f"\nNombres de columnas:")
print(df.columns.tolist())
print(f"\nTipos de datos:")
print(df.dtypes)
```
